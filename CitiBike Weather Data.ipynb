```python
import ast

with open("temp_data.txt") as f:
    text = f.read()
    
d = ast.literal_eval(text)
```

```python
avg_temps = [item for item in d['results'] if item['datatype']=='TAVG']
```

```python
date_temp = [item['date'] for item in avg_temps]
```

```python
temps = [item['value'] for item in avg_temps]
```

```python
df_temp = pd.DataFrame()
df_temp['Date'] = [datetime.strptime(d, "%Y-%m-%dT%H:%M:%S") for d in date_temp] 
df_temp['Average Tempurature'] = [float(v)/10.0 for v in temps]
```

```python
df['started_at'] = pd.to_datetime(df['started_at'], dayfirst = True)
df['Date'] = pd.to_datetime(df['started_at'], format='%Y-%m-%d').dt.date 
```

```python
df.tail(5)
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df_temp['Date'] = pd.to_datetime(df_temp['Date'])
```

```python
%%time
df_merged = df.merge(df_temp, how = 'left', on = 'Date', indicator = True) 
```
