Step 4: Create a column with the count of trips per day and merge it with the weather dataframe.

```python
df_group = pd.DataFrame(df_temp.groupby(['date'])['ride_id'].count()).reset_index()
```

```python
df_group.rename(columns = {'ride_id':'bike_rides_daily'}, inplace = True)
df = df_temp.merge(df_group, on = "date", how = 'outer', indicator = True)
print(df['_merge'].value_counts(dropna = False))
print("Shape of January 1st is", df[df['date'] == '2018-01-01'].shape) # Check 
print("Shape of January 2nd is", df[df['date'] == '2018-01-02'].shape) # Second check 

df_fuse = df.set_index('date')

print(df_fuse.columns)
```

```python
plt.figure(figsize=(10,5))
plt.title('Temperature per trips')
plt.plot(df_fuse['average_tempurature'])
plt.plot(df_fuse['bike_rides_daily'], color = 'red')
plt.show()
```

Step 5: Create a line chart of bike trip counts and temperatures plotted on a dual axis. 

```python
# Create figure and axis objects with subplots()

fig,ax = plt.subplots(figsize=(10, 5))

# Make a plot of the bike rides
ax.plot(df_fuse['bike_rides_daily'],color = "red")

# Here is where we create the dual axis - initialize a twin object for two different y-axis on the sample plot
ax2 = ax.twinx()

# Make a plot with the temperatues on the y-axis using second axis object
ax2.plot(df_fuse["average_tempurature"], color = "blue")
plt.show()
```

```python
# Create figure and axis objects with subplots() and details

fig,ax = plt.subplots(figsize=(10, 5))

# Make a plot of the bike rides
ax.plot(df_fuse['bike_rides_daily'],color = "navy")
ax.set_xlabel("Year 2022", fontsize = 14) # set x-axis label
ax.set_ylabel("Bike rides daily",color = "navy", fontsize = 14) # set y-axis label

# Here is where we create the dual axis - initialize a twin object for two different y-axis on the sample plot
ax2 = ax.twinx()

# Make a plot with the temperatues on the y-axis using second axis object
ax2.plot(df_fuse["average_tempurature"], color = "red")
ax2.set_ylabel("Average temperatures", color = "red",fontsize=14)
plt.title('Temperature and trips in 22', fontsize = 18)
plt.show()
```
